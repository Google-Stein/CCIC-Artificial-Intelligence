```python
from bs4 import BeautifulSoup
import requests

def extract_all_books(html):
    soup = BeautifulSoup(html, "html.parser")
    
    book_entries = soup.find_all("li", class_="searchResultItem")
    
    books = []

    for book in book_entries:
        title = book.find("h3", class_="booktitle").text.strip()
        
        author = book.find("span", class_="bookauthor").find("a").text.strip()
        
        year_published = book.find("span", class_="resultDetails").find(text=lambda t: "First published" in t).strip().split(" ")[-1]
        
        logs_text = book.find("div", class_="details").text
        logs = int([word for word in logs_text.split() if word.isdigit()][-1])
      
        books.append({
            "title": title,
            "author": author,
            "year_published": year_published,
            "logs": logs
        })
    
    return books

response = requests.get("https://openlibrary.org/trending/forever") 
html_content = response.text

all_books = extract_all_books(html_content)

for book in all_books:
    print(book)
```
