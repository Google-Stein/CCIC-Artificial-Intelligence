```python
IvyGrad = 0
Total = 0
base_url = "https://en.wikipedia.org"

def findEducation(page):
    url = page
    fetchedResponse = requests.get(url)
    soup = BeautifulSoup(fetchedResponse.content, 'html.parser')

    alma_mater_row = soup.find('th', text="Alma mater")

    if alma_mater_row:
        alma_mater_data = alma_mater_row.find_next_sibling('td')
        if alma_mater_data:
            alma_mater_links = alma_mater_data.find_all('a')
            alma_mater_names = [link.get_text() for link in alma_mater_links]
            print("Alma Mater:", alma_mater_names)
            return alma_mater_names
        else:
            print("Alma Mater data not found.")
            return []
    else:
        print("Alma Mater row not found.")
        return []

links = [a['href'] for a in soup.find_all('a', href=True)]

# Iterate through each link
for link in links:
    # Check if the link is relative or absolute
    if link.startswith("http"):
        full_url = link  # Use the full URL as is
    else:
        full_url = f"{base_url}{link}"  # Append base URL to relative path
    
    # Fetch the content of the linked page
    try:
        response = requests.get(full_url)
        response.raise_for_status()  # Raise an error for bad status codes
        
        # Call findEducation to extract alma mater from the linked page
        alma_mater_names = findEducation(full_url)
        
        ivy = ["Harvard University", "Yale University", "Princeton University", "Cornell University", "Dartmouth College", "Columbia University", "Brown Univsersity", "University of Pennsylvania"]

        for name in ivy:
            if name in alma_mater_names:
                IvyGrad += 1
                Total += 1
            else:
                Total += 1
    except requests.exceptions.RequestException as e:
        print(f"Failed to fetch {full_url}: {e}")

print(IvyGrad, Total)
```
