```python
def get_best_fit(df, x, y):
    
    m = 0
    b = 0
    
    #Calculate Residuals
    res = 0
    for count in range(len(df)):
        m = m + 1
        b = 0
        
        res = 0
        
        for i in range(len(df)):
            pred_y = (m * df.loc[i][x]) + b
            act_y = df.loc[i][y]
            temp = act_y - pred_y
            temp2 = temp ** 2
            res = res + temp2
        print(res)
    
    return(m, b, res)
```

```python
def determination(df, x, y):
    y_mean = df[y].mean()
    
    ss_total = ((df[y] - y_mean) ** 2).sum()
    
    x_mean = df[x].mean()
    slope = ((df[x] - x_mean) * (df[y] - y_mean)).sum() / ((df[x] - x_mean) ** 2).sum()
    intercept = y_mean - slope * x_mean
    
    y_pred = intercept + slope * df[x]
    
    ss_residual = ((df[y] - y_pred) ** 2).sum()
    
    r_squared = 1 - (ss_residual / ss_total)
    
    return r_squared
```

```python
determination(fullConDf, 'budget', 'revenue')
```
