```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

eng = pd.read_json('english_movies.json')
french = pd.read_json('french_movies.json')
```

```python
eng.head()
```

```python
fr = french.drop_duplicates(subset=['id'])
fr
```

```python
conDf = pd.concat([fr, eng], ignore_index=True)
```

```python
sns.regplot(data=conDf, x='budget', y='revenue')
```
